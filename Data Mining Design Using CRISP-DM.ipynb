```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from scipy import stats
```

```python
# Add dataset
midterm_grades = np.array([72, 50, 81, 74, 94, 86, 59, 83, 65, 33, 88, 81])
final_grades = np.array([84, 63, 77, 78, 90, 75, 49, 79, 77, 52, 74, 90])
```

```python
# Plotting the data
plt.scatter(midterm_grades, final_grades, color='blue', label='Data Points')
plt.xlabel('Midterm Grade Scores')
plt.ylabel('Final Grade Scores')
plt.title('Midterm vs Final Exam Grade Scores (%)')

# Fitting a linear regression model
X = sm.add_constant(midterm_grades)  # Adds a constant term to the predictor
model = sm.OLS(final_grades, X).fit()
predictions = model.predict(X)  # Predictions from the model

# Plotting the regression line
plt.plot(midterm_grades, predictions, color='red', label='Regression Line')
plt.legend()
plt.show()

# Print out the summary of the regression
print(model.summary())

# Extracting key metrics
slope = model.params[1]
intercept = model.params[0]
r_squared = model.rsquared
p_value = model.pvalues[1]

print(f'Slope (m): {slope}')
print(f'Intercept (b): {intercept}')
print(f'R-squared: {r_squared}')
print(f'P-value: {p_value}')
```

```python
# Prediction
midterm_grade = 86
predicted_final_grade = model.predict([1, midterm_grade])
print(f'Predicted Final Exam Grade for midterm grade of {midterm_grade}: {predicted_final_grade[0]}')
```
